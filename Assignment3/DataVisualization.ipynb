## Trực quan với từng group test

```python
a=[]
group_test=0
for a in range(0,78,6):
  labels = ['50','100','200','500','1000','2000']
  x = np.arange(len(labels))  # the label locations
  width = 0.25  # the width of the bars

  fig, ax = plt.subplots()

  # rects1 = ax.bar(x - width/2, value_arr[1][a:a+6], width, label='GA_50')

  # rects2 = ax.bar(x -width/4, value_arr[1][a:a+6], width, label='GA_100')
  # rects3 = ax.bar(x +width/2, value_arr[2][a:a+6], width, label='GA_200')
  rects4 = ax.bar(x +width, value_arr[3][a:a+6], width, label='GA_300')
  rects5 = ax.bar(x +width*1.25, value_arr[4][a:a+6], width, label='GA_2Minute')
  rects6 = ax.bar(x +width*1.5, value_arr[5][a:a+6], width, label='GA_3Minute')
  rects6 = ax.bar(x +width*1.75, value_arr[6][a:a+6], width, label='OR')
  print("Group: "+str(lst[group_test]))
  # Add some text for labels, title and custom x-axis tick labels, etc.
  ax.set_ylabel('Value')
  ax.set_title('Value by size testcase')
  ax.set_xticks(x)
  ax.set_xticklabels(labels)
  ax.legend()
  group_test+=1
  fig.tight_layout()
  a=plt.show()
```
